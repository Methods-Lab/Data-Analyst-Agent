## 1.6 Summary statistics (numeric columns)

`.describe()` gives count, mean, std, min, quartiles, and max for every numeric column. This is the single most important first-look command in pandas.

```python
df.describe()
```

## 1.7 Categorical columns — value counts

For text/category columns we want to know how many distinct values exist and which are most common.

```python
for col in df.select_dtypes(include='object').columns[:4]:
    print(f'\n--- {col} ---')
    print(df[col].value_counts().head(5))
```

## 1.8 Missing-value scan

Knowing missing-value counts up front decides our preprocessing strategy in Step 2.

```python
missing = df.isna().sum()
missing_pct = (df.isna().mean() * 100).round(2)
pd.DataFrame({'missing_count': missing, 'missing_pct': missing_pct})\
  .sort_values('missing_count', ascending=False).head(10)
```
